```python
import matplotlib.pyplot as plt

n = []
t = []
with open('mult.bench') as f:
    for line in f:
        s = line.split(' ')
        n.append(int(s[0]))
        t.append(float(s[1]))

plt.plot(n, t)
plt.xlabel('N')
plt.ylabel('Time in seconds')
plt.title('Random NxN matrix multiplication time')
plt.savefig('mult.jpg')
plt.show()
```
